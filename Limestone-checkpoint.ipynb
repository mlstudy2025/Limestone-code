```python
# Classify Overburden Thickness
def classify_overburden(thickness):
    if thickness < 10:
        return "Low Overburden"
    elif 10 <= thickness <= 25:
        return "Moderate Overburden"
    else:
        return "High Overburden"

df["Overburden Class"] = df["Overburden thickness(m)"].apply(classify_overburden)

# Define LSI Classification based on Industry Standards
def classify_lsi(row):
   if (row['CaO'] >= 49 and row['Fe2O3'] <= 1.5 and 
        row['Al2O3'] <= 1.5 and row['MgO'] <= 1.2 and row['SiO2'] <= 8 and row['LOI'] >= 42 and 
        row['Impurity Index (%)'] <= 15 and row["CaO/SiO2 Ratio"] >= 6 and row['Limestone thickness (m)'] >= 15 and row["Overburden Class"] == "Low Overburden"):
        return "High LSI"  # High Suitability
    
      # Moderate Suitability (Uses Only 2-3 Key Variables)
   elif row['CaO'] >= 45 and row['Limestone thickness (m)'] >= 7:
        return "Moderate LSI"    # Moderate Suitability
    
   else:
        return "Low LSI"  # Low Suitability

# Apply function to classify each row in the DataFrame
df['LSI Classification'] = df.apply(classify_lsi, axis=1)

# Display the updated dataset
df[["SiO2", "Al2O3", "Fe2O3", "MgO", "CaO", "LOI", "CaO/SiO2 Ratio", "Impurity Index (%)",
          "Limestone thickness (m)", "Overburden thickness(m)", "Overburden Class", "LSI Classification"]]

# Save the updated dataset
#df.to_excel("LSI_Classification_with_Chemistry.xlsx", index=False)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, make_scorer

# Load your dataset (replace with actual path)
df = pd.read_csv(r"C:\Users\HP\Desktop\gis_output.csv")

# Define features and target variable
X = df.drop(columns=["LSI"])  # Predictors (Assumed categorical target)
y = df["LSI"]  # Target variable (Categorical: "High", "Moderate", "Low")

# Standardize features to improve model performance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define K-Fold Cross-Validation (5 folds)
kf = KFold(n_splits=5, shuffle=True)

# Initialize Random Forest Classifier with optimized hyperparameters
model = RandomForestClassifier(
    n_estimators=200,       # More trees for better learning
    max_depth=10,           # Prevent overfitting
    min_samples_split=5,    # Controls tree complexity
    min_samples_leaf=2,     # Prevents very small splits
    n_jobs=-1, 
    random_state=42             # Faster computation
)

# Perform Cross-Validation with Accuracy as the metric
cv_scores = cross_val_score(model, X_scaled, y, cv=kf, scoring="accuracy", n_jobs=-1)

# Print results
print(f"Cross-validation accuracy scores: {cv_scores}")
print(f"Mean Accuracy Score: {np.mean(cv_scores):.3f}")
print(f"Standard deviation of scores: {np.std(cv_scores):.3f}")
```

```python
import numpy as np
# Set font to Arial
plt.rcParams['font.family'] = 'Arial'

df = pd.read_csv(r"C:\Users\HP\Desktop\gis_output2.csv") 

# Define confidence levels based on probability
def classify_confidence(prob):
    if prob >= 90:
        return "High"
    elif prob >= 70:
        return "Moderate"
    else:
        return "Low"

df["Confidence_Level"] = df["Probability (%)"].apply(classify_confidence)

# Count boreholes in each LSI and Confidence Level
lsi_confidence_table = df.groupby(["LSI_Quality", "Confidence_Level"]).size().unstack()

# Plot bar chart
lsi_confidence_table.plot(kind="bar", stacked=True, figsize=(8, 5), colormap="viridis")
plt.xlabel("LSI Quality")
plt.ylabel("Number of Boreholes")
plt.title("LSI Quality vs. Confidence Level")
plt.legend(title="Confidence Level")
#Save first to PNG
plt.savefig('SI Quality vs. Confidence Level_temp.png', dpi=300, bbox_inches='tight')

# Open PNG and re-save as TIFF with compression
img = Image.open('SI Quality vs. Confidence Level_temp.png')
img.save('SI Quality vs. Confidence Level_final.tiff', compression='tiff_lzw')

# Optional: Remove temp png
import os
os.remove('SI Quality vs. Confidence Level_temp.png')
plt.show()
```
